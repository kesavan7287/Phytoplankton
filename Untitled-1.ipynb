```python
import pandas as pd
import osgb
import numpy as np
import os
```

```python
data = pd.read_csv('pythobook.csv')
data.head()
```

```python
def grid_ref_to_coords(grid_ref):
    try:
        (easting, northing) = osgb.parse_grid(grid_ref)
        (lat, lon) = osgb.grid_to_ll(easting, northing)
        return lat, lon
    except:
        return -1, -1

grid_ref_to_coords("SW85134212")
```

```python
data['GridReference'].unique()
```

```python
ret = pd.DataFrame()
ret['Grid'] = data['GridReference'].unique()
```

```python
ret.head()
```

```python
def long_from_osgb(x):
    return grid_ref_to_coords(x)[1]

def lat_from_osgb(x):
    return grid_ref_to_coords(x)[0]
```

```python
ret['Longitude'] = ret['Grid'].apply(lambda x: long_from_osgb(x))
ret['Latitude'] = ret['Grid'].apply(lambda x: lat_from_osgb(x))
```

```python
path = r'../Phytoplankton'
ret.to_csv(os.path.join(path,r'coords.csv'))
```
